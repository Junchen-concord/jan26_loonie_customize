# Production model training metadata

This notebook loads the production AutoGluon model and prints its training metadata (bagging, stack levels, presets, etc.).

```python
from __future__ import annotations

import os
import sys
from autogluon.tabular import TabularPredictor

PROJECT_ROOT = "/Users/starsrain/jan2026_concord/jan2026_loonie_customize"
SRC_DIR = os.path.join(PROJECT_ROOT, "src")
if SRC_DIR not in sys.path:
    sys.path.append(SRC_DIR)

from config.config import REDZONE_MODEL_FILE_PATH_V2

MODEL_DIR = REDZONE_MODEL_FILE_PATH_V2
print("Production model path:", MODEL_DIR)

predictor = TabularPredictor.load(MODEL_DIR, require_py_version_match=False)

best_model = getattr(predictor, "model_best", None)
if not best_model:
    try:
        best_model = predictor.get_model_best()
    except Exception:
        best_model = None
print("Best model:", best_model)

print("\nFit summary (keys):")
fit_summary = predictor.fit_summary(show_plot=False)
print(sorted(fit_summary.keys()))

fit_meta = fit_summary.get("fit_meta", {}) if isinstance(fit_summary, dict) else {}
print("\nFit meta:")
print({
    "num_bag_folds": fit_meta.get("num_bag_folds"),
    "num_bag_sets": fit_meta.get("num_bag_sets"),
    "num_stack_levels": fit_meta.get("num_stack_levels"),
    "presets": fit_meta.get("presets"),
})

print("\nModel info:")
info = predictor.info()
print({"version": info.get("version"), "stack_levels": info.get("stack_levels"), "bag_folds": info.get("bag_folds")})
```
